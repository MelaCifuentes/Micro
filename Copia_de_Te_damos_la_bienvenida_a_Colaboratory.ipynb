Con Colab, puedes aprovechar toda la potencia de las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código de abajo utiliza <strong>NumPy</strong> para generar datos aleatorios y <strong>Matplotlib</strong> para visualizarlos. Puedes importar tus propios datos a los cuadernos de Colab desde tu cuenta de Google Drive, incluidas las hojas de cálculo, y también desde GitHub y muchas fuentes más. Para obtener más información sobre cómo importar datos y cómo se puede usar Colab en la ciencia de datos, consulta los enlaces que aparecen en la sección <a href="#working-with-data">Trabajar con datos</a> más abajo.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt

# Datos
meses = ["jun-25", "jul-25", "ago-25", "sept-25", "oct-25", "nov-25", "dic-25", "ene-26", "feb-26"]
liquidez = [33.27, 36.57, 37.57, 36.72, 37.25, 40.04, 41.09, 42.14, 42.12]

# Crear gráfico lineal
plt.figure(figsize=(9,5))
plt.plot(meses, liquidez, marker='o', linestyle='-', color='blue')

# Etiquetas de valores
for i, valor in enumerate(liquidez):
    plt.text(i, valor, f"{valor:.2f}%", ha='center', fontsize=9)

# Títulos y etiquetas
plt.title("Evolución del Indicador de Liquidez (%)")
plt.xlabel("Periodo")
plt.ylabel("Liquidez (%)")

# Rotar etiquetas del eje X
plt.xticks(rotation=45)

# Mostrar cuadrícula
plt.grid(True)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# =========================
# 📊 DATOS
# =========================
oficinas = [
    "Matriz","Latacunga","Guaranda","Riobamba","Milagro",
    "Babahoyo","Agencia Sur","Baños","Quito","Cotocollao",
    "Ter. Terrestre","Agencia Norte","Quevedo"
]

morosidad = [
    10.37,7.13,5.83,7.11,10.82,
    9.65,9.07,7.14,15.38,25.66,
    7.45,4.64,17.12
]

# =========================
# 📌 INDICADORES
# =========================
promedio = 9.63
desv_std = np.std(morosidad)

banda_sup = promedio + desv_std
banda_inf = promedio - desv_std

# =========================
# 📊 GRÁFICO
# =========================
plt.figure(figsize=(12,6))

# Línea principal
plt.plot(oficinas, morosidad,
         marker='o',
         color='#6A0DAD',
         linewidth=2,
         label='Morosidad (%)')

# Promedio
plt.axhline(promedio,
            color='green',
            linestyle='--',
            linewidth=2,
            label=f'Promedio ({promedio:.2f}%)')

# Bandas de control
plt.axhline(banda_sup,
            color='red',
            linestyle='--',
            label=f'+1 Desv. ({banda_sup:.2f}%)')

plt.axhline(banda_inf,
            color='orange',
            linestyle='--',
            label=f'-1 Desv. ({banda_inf:.2f}%)')

# =========================
# 🏷️ ETIQUETAS
# =========================
for i, v in enumerate(morosidad):
    plt.text(i, v + 0.5,
             f"{v:.2f}%",
             ha='center',
             fontsize=9,
             fontweight='bold')

# =========================
# 🎨 ESTÉTICA
# =========================
plt.title("Morosidad por Oficina con Bandas de Control",
          fontsize=16,
          fontweight='bold',
          color='#6A0DAD')

plt.xlabel("Oficinas", fontweight='bold')
plt.ylabel("% Morosidad", fontweight='bold')

plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Datos
segmentos = ["Consumo", "Microcrédito", "Productivo Pymes", "Refinanciados", "Inmobiliario"]

operaciones = np.array([3955773, 1472450, 414500, 165883, 200000])
concentracion = np.array([63.71, 23.72, 6.68, 2.97, 3.22])

# Función de riesgo
def riesgo(valor):
    if valor >= 50:
        return "CRÍTICO", "#B71C1C"  # rojo oscuro
    elif valor >= 20:
        return "ALTO", "#EF6C00"     # naranja
    elif valor >= 5:
        return "MEDIO", "#FDD835"    # amarillo
    else:
        return "BAJO", "#2E7D32"     # verde

colores = []
niveles = []

for val in concentracion:
    r, c = riesgo(val)
    niveles.append(r)
    colores.append(c)

# Crear figura
plt.figure(figsize=(12,4))

x = np.arange(len(segmentos))

# Barras
bars = plt.bar(x, concentracion, color=colores, edgecolor="black")

# Etiquetas dentro de barras
for i in range(len(segmentos)):
    plt.text(i, concentracion[i]/2,
             f"{concentracion[i]:.2f}%\n{operaciones[i]:,.0f}\n{niveles[i]}",
             ha='center',
             va='center',
             fontsize=11,
             fontweight='bold',
             color="white")

# Ejes
plt.xticks(x, segmentos, fontsize=12)
plt.ylabel("% Concentración", fontsize=13)

# Título estilo dashboard
plt.title("Matriz de Riesgo – Cartera por Segmento",
          fontsize=16,
          fontweight='bold',
          color="#EB27F5")

# Grid suave
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Datos
segmentos = ["Consumo", "Microcrédito", "Productivo Pymes", "Refinanciados", "Inmobiliario"]

operaciones = np.array([3955773, 1472450, 414500, 165883, 200000])
concentracion = np.array([63.71, 23.72, 6.68, 2.97, 3.22])

# Matriz (1 fila)
data = concentracion.reshape(1, -1)

# Función de riesgo → colores personalizados
def color_riesgo(val):
    if val >= 50:
        return 4  # CRÍTICO
    elif val >= 20:
        return 3  # ALTO
    elif val >= 5:
        return 2  # MEDIO
    else:
        return 1  # BAJO

riesgo_matrix = np.array([[color_riesgo(v) for v in concentracion]])

# Colores tipo semáforo
cmap = plt.cm.get_cmap("RdYlGn_r", 4)

plt.figure(figsize=(12,3))

# Heatmap
plt.imshow(riesgo_matrix, cmap=cmap, aspect="auto", vmin=1, vmax=4)

# Etiquetas
plt.xticks(range(len(segmentos)), segmentos, fontsize=12)
plt.yticks([0], ["Riesgo"], fontsize=12)

# Texto dentro de cada celda
for i in range(len(segmentos)):
    # Determinar nivel
    if concentracion[i] >= 50:
        nivel = "CRÍTICO"
        color_texto = "white"
    elif concentracion[i] >= 20:
        nivel = "ALTO"
        color_texto = "black"
    elif concentracion[i] >= 5:
        nivel = "MEDIO"
        color_texto = "black"
    else:
        nivel = "BAJO"
        color_texto = "black"

    plt.text(i, 0,
             f"{concentracion[i]:.2f}%\n{operaciones[i]:,.0f}\n{nivel}",
             ha='center',
             va='center',
             fontsize=11,
             fontweight='bold',
             color=color_texto)

# Barra de color
cbar = plt.colorbar(ticks=[1,2,3,4])
cbar.set_ticklabels(["BAJO", "MEDIO", "ALTO", "CRÍTICO"])
cbar.set_label("Nivel de Riesgo")

# Título
plt.title("Mapa de Calor – Matriz de Riesgo de Cartera",
          fontsize=16,
          fontweight='bold',
          color="#EB27F5")

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import squarify

# Datos
segmentos = ["Consumo", "Microcrédito", "Productivo Pymes", "Refinanciados", "Inmobiliario"]

concentracion = [63.71, 23.72, 6.68, 2.97, 3.22]

operaciones = [3955773, 1472450, 414500, 165883, 200000]

# Función de riesgo
def riesgo(val):
    if val >= 50:
        return "CRÍTICO", "#C62828"
    elif val >= 20:
        return "ALTO", "#EF6C00"
    elif val >= 5:
        return "MEDIO", "#FDD835"
    else:
        return "BAJO", "#2E7D32"

colores = []
labels = []

for seg, conc, op in zip(segmentos, concentracion, operaciones):
    nivel, color = riesgo(conc)
    colores.append(color)
    labels.append(f"{seg}\n{conc:.2f}%\n{op:,.0f}\n{nivel}")

# Crear figura
plt.figure(figsize=(12,6))

# Treemap
squarify.plot(
    sizes=concentracion,
    label=labels,
    color=colores,
    alpha=0.9,
    text_kwargs={'fontsize':11, 'fontweight':'bold'}
)

# Título estilo dashboard
plt.title("Treemap de Cartera – Concentración y Riesgo",
          fontsize=16,
          fontweight='bold',
          color="#EB27F5")

plt.axis("off")

plt.tight_layout()
plt.show()
```

```
import plotly.graph_objects as go
import pandas as pd

# Datos
df = pd.DataFrame({
    "Segmento": ["Consumo", "Microcrédito", "Productivo Pymes", "Refinanciados", "Inmobiliario"],
    "Operaciones": [3955773, 1472450, 414500, 165883, 200000]
})

# Ordenar de mayor a menor
df = df.sort_values("Operaciones", ascending=False)

# % acumulado
df["% acumulado"] = df["Operaciones"].cumsum() / df["Operaciones"].sum() * 100

# 🎨 colores
bar_color = "#00C2FF"
line_color = "#6A0DAD"

# 📊 crear gráfico
fig = go.Figure()

# 🔹 Barras
fig.add_trace(go.Bar(
    x=df["Segmento"],
    y=df["Operaciones"],
    name="Operaciones",
    marker_color=bar_color,
    text=[f"{v:,.0f}" for v in df["Operaciones"]],
    textposition="outside"
))

# 🔹 Línea acumulada
fig.add_trace(go.Scatter(
    x=df["Segmento"],
    y=df["% acumulado"],
    name="% Acumulado",
    yaxis="y2",
    mode="lines+markers+text",
    line=dict(color=line_color, width=3),
    text=[f"{v:.2f}%" for v in df["% acumulado"]],
    textposition="top center"
))

# 📌 Layout
fig.update_layout(
    title={
        "text": "<b>📈 Análisis de Pareto - Concentración de Operaciones</b>",
        "x": 0.5,
        "font": {"size": 24, "color": "#6A0DAD"}
    },

    xaxis=dict(title="Segmento"),

    yaxis=dict(
        title="Operaciones"
    ),

    yaxis2=dict(
        title="% Acumulado",
        overlaying="y",
        side="right",
        range=[0, 100]
    ),

    legend=dict(x=0.8, y=1.15),

    margin=dict(t=80, b=60)
)

fig.show()
```

```
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# =========================
# 📊 DATOS
# =========================
df = pd.DataFrame({
    "Segmento": ["Consumo", "Microcrédito", "Productivo Pymes", "Refinanciados", "Inmobiliario"],
    "Operaciones": [3955773, 1472450, 414500, 165883, 200000],
    "Concentracion": [63.71, 23.72, 6.68, 2.97, 3.22]
})

df = df.sort_values("Operaciones", ascending=False)

# Pareto
df["% acumulado"] = df["Operaciones"].cumsum() / df["Operaciones"].sum() * 100

# =========================
# 🎨 COLORES
# =========================
colores = ["#00C2FF", "#00E5A8", "#FFD166", "#FF6B6B", "#845EF7"]
morado = "#6A0DAD"
fondo = "#F4F6F9"

# =========================
# 📊 SUBPLOTS
# =========================
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Embudo de Operaciones", "Análisis de Pareto"),
    specs=[[{"type": "funnel"}, {"secondary_y": True}]]
)

# =========================
# 🔻 FUNNEL
# =========================
fig.add_trace(go.Funnel(
    y=df["Segmento"],
    x=df["Operaciones"],
    textinfo="none",
    marker=dict(color=colores)
), row=1, col=1)

x_max = df["Operaciones"].max()

for _, row in df.iterrows():

    if row["Segmento"] in ["Consumo", "Microcrédito"]:
        fig.add_annotation(
            x=row["Operaciones"]/2,
            y=row["Segmento"],
            text=f"<b>{row['Operaciones']:,.0f}</b><br>{row['Concentracion']:.2f}%",
            showarrow=False,
            font=dict(size=18, color="white"),
            xref="x1", yref="y1"
        )

    elif row["Operaciones"] < x_max * 0.12:
        # 🔥 etiqueta fuera con línea
        fig.add_annotation(
            x=row["Operaciones"] * 1.05,
            y=row["Segmento"],
            text=f"<b>{row['Operaciones']:,.0f}</b><br>{row['Concentracion']:.2f}%",
            showarrow=True,
            arrowhead=2,
            ax=40,
            ay=0,
            font=dict(size=14, color="#111111"),
            xref="x1", yref="y1"
        )

    else:
        fig.add_annotation(
            x=row["Operaciones"]/2,
            y=row["Segmento"],
            text=f"<b>{row['Operaciones']:,.0f}</b><br>{row['Concentracion']:.2f}%",
            showarrow=False,
            font=dict(size=16, color="white"),
            xref="x1", yref="y1"
        )

# =========================
# 📈 PARETO
# =========================
fig.add_trace(go.Bar(
    x=df["Segmento"],
    y=df["Operaciones"],
    marker_color="#00C2FF",
    name="Operaciones",
    text=[f"{v:,.0f}" for v in df["Operaciones"]],
    textposition="outside"
), row=1, col=2, secondary_y=False)

fig.add_trace(go.Scatter(
    x=df["Segmento"],
    y=df["% acumulado"],
    mode="lines+markers+text",
    name="% acumulado",
    line=dict(color=morado, width=3),
    text=[f"{v:.2f}%" for v in df["% acumulado"]],
    textposition="top center"
), row=1, col=2, secondary_y=True)

# línea 80%
fig.add_hline(y=80, line_dash="dash", line_color="red", row=1, col=2)

# =========================
# 🎨 ESTILO DASHBOARD
# =========================
fig.update_layout(
    title={
        "text": "<b>📊 Dashboard de Concentración de Cartera</b>",
        "x": 0.5,
        "font": {"size": 26, "color": morado}
    },
    plot_bgcolor=fondo,
    paper_bgcolor=fondo,
    font=dict(size=14, color="#333"),
    height=600,
    margin=dict(t=80, l=60, r=60, b=50)
)

fig.update_yaxes(title_text="Operaciones", row=1, col=2, secondary_y=False)
fig.update_yaxes(title_text="% Acumulado", row=1, col=2, secondary_y=True)

fig.show()

# =========================
# 📦 EXPORTAR A HTML
# =========================
fig.write_html("dashboard_concentracion.html")
```
